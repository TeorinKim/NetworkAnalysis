```python
#raw_sentence=list(voc_all['voc'])
raw_sentence=list(sen_temp)
```

```python
dic_words = list(dic_all['단어'])
```

```python
dic_word6=dic_all[dic_all['단어길이']>5]
dic_word5=dic_all[dic_all['단어길이']==5]
dic_word4=dic_all[dic_all['단어길이']==4]
dic_word3=dic_all[dic_all['단어길이']==3]
dic_word2=dic_all[dic_all['단어길이']==2]
dic_word1=dic_all[dic_all['단어길이']==1]
```

```python
dic6 = list(dic_word6['단어'])
dic5 = list(dic_word5['단어'])
dic4 = list(dic_word4['단어'])
dic3 = list(dic_word3['단어'])
dic2 = list(dic_word2['단어'])
dic1 = list(dic_word1['단어'])
```

```python
trash=[]

voc_result=[]

for sentence in raw_sentence:
    temp1=sentence
    temp_word=[]
    for word in dic6:
        if temp1.find(word)> -1 :
            temp1=temp1.replace(word, " ")
            temp_word.append(word)
    for word in dic5:
        if temp1.find(word)> -1 :
            temp1=temp1.replace(word, " ")
            temp_word.append(word)
    for word in dic4:
        if temp1.find(word)> -1 :
            temp1=temp1.replace(word, " ")
            temp_word.append(word)
    for word in dic3:
        if temp1.find(word)> -1 :
            temp1=temp1.replace(word, " ")
            temp_word.append(word)
    for word in dic2:
        if temp1.find(word)> -1 :
            temp1=temp1.replace(word, " ")
            temp_word.append(word)
    for word in dic1:
        if temp1.find(word)> -1 :
            temp1=temp1.replace(word, " ")
            temp_word.append(word)

    voc_result.append(temp_word)
    trash.append(temp1)

            
```

```python
df_voc=pd.DataFrame({'voc_basement': raw_sentence , 'voc_words':voc_result, 'trash': trash})
```
