```python
recode_result=[]

for j in range(len(voc_all)):
    temp_words=voc_all['result_all'][j]
    temp_area=str(voc_all['area'][j])
    temp_result=[]
    for word in temp_words:
        temp0=word
        for temp1, temp2, temp3, pre1, pre2, pre3, post, del1, del2 in recode_loc_list:
            temp1=str(temp1)
            temp2=str(temp2)
            pre1=str(pre1)
            pre2=str(pre2)
            pre3=str(pre3)
            post=str(post)
            del1=str(del1)
            del2=str(del2)
            temp3 = temp1 + " " +temp2
        
            if (temp2 == 'nan' and temp_area.find(temp1) > -1 ) | (temp3 == temp_area):
                if word == pre1 and pre1 != 'nan':
                    temp0 = post
                elif word == pre2 and pre2 != 'nan':
                    temp0 = post
                elif word == pre3 and pre3 != 'nan':
                    temp0 = post
                elif word == del1 and del1 != 'nan':
                    temp0 = ""
                elif word == del2 and del2 != 'nan':
                    temp0 = ""
        
        temp_result.append(temp0)
    recode_result.append(temp_result)
```

```python
def cleaning(x):
    if str(x) == "인천광역시 남구":
        return str(x).replace("남구", "미추홀구")
    elif str(x) == "제주특별자치도 제주시":
        return str(x).replace("제주시", "제주(시내동)")
    elif str(x) == "제주특별자치도 서귀포시":
        return str(x).replace("서귀포시", "서귀포(시내동)")
    return str(x)
```
